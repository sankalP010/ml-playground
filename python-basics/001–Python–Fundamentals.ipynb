# **2. Data Types in Python**

```python
# Integer
print(10)
# Max - 1 * 10^308
print(1e309)
type(10)
```

```python
# Floating Number
print(2.9)
# Max: 1.7 * 10^308
print(1.7e309)
type(1e308)
```

```python
# Character String
print("sankalp")
print('a')
type('a')
```

```python
# Boolean
print(True)
print(False)
type(True)
```

```python
# Complex Numbers: only j signifies imaginary value
print(4 + 7j)
type(4 + 7j)
```

```python
# Array is called Lists in Python
print([1, 2, 3, 4, 5])
type([1, 2])
```

```python
# Tuple
print((1, 2, 3, 4, 5))
type((1, 2, 3))
```

```python
# Sets
print({1, 2, 3, 4, 5})
type({1, 2, 3, 4})
```

```python
# Dictionary
print({'name': 'Sankalp', 'age': 24, 'place': 'jabalpur'})
type({'name': 'Sankalp', 'id': 10})
```

# **3. Variables in Python**

```python
# Variables are containers which stores information
# for future use
a = 7
print(a)
```

```python
# Python supports Dynamic Typing
a = 7
print(a)
## Static Typing: java
## int a = 7;
```

```python
# Python supports Dynamic Binding
a = 10
print(a)
type(a) # Dynamic Binding
a = "sankalp"
print(a)
type(a)
## Static Binding
## int a = 10;
## Now, data type of a cannot be changed.
```

```python
# Other ways of Declaration of Variable in Python
a = 10
b = 20
c = 30
print(a, b, c)
## Way 1
d, e, f = 40, 50, 60
print(d, e, f)
## Way 2
g = h = i = 70
print(g, h, i)
```

# **7. Literals**

```python
# Literals are value stored in a variable
```

```python
# Binary Literal
binary = 0b1010
# Decimal Literal
decimal = 100;
# Octal Literal
octal_value = 0o310
# Hexadecimal Literal
hexa_value = 0x12c
# Complex Literal
comp_1 = 3 + 4j
comp_2 = 7j
# float Literal
float_1 = 5.89
float_2 = 1e4
float_3 = 1.5e-3
print(binary, decimal, octal_value, hexa_value)
print(comp_1.real, comp_2.real, comp_1.imag, comp_2.imag)
print(float_1, float_2, float_3)
```

```python
# String Literal
str_1 = "This is a string literal"
str_2 = 'This is a string literal'
char = 'C'
multi_str = """This is a multi-line
string literal"""
unicode = u"\U0001f600\U0001F606\U0001F923"
raw_str = r"raw \n string"
print(unicode)
print(char)
print(raw_str)
print(str_1)
print(str_2)
print(multi_str)
```

```python
# Boolean Literal
val_1 = False + 7
val_2 = True + 7
print(val_1, val_2)
```

```python
# Special Literal -> None
a = None
print(a)
```
